```python
import camb
import numpy as np 
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.integrate import quad

print(f"version: (camb._version_)")
```

```python
H0=67.4
omega_m=0.315
omega_ch2=0.12
omega_bh2=0.0224
ns=0.965
tau=0.054
```

```python
pars=camb.CAMBparams();
pars.set_cosmology(H0=H0,ombh2=omega_bh2,omch2=omega_ch2,tau=tau);
```

```python
results = camb.get_results(pars)
```

```python
pars.set_dark_energy(w=-1);
```

```python
z=np.linspace(0,10000,100000)
```

```python
a=1/(1+z)
print(a)
```

```python
densK=results.get_Omega('K',z)
denscdm=results.get_Omega('cdm',z)
densb=results.get_Omega('baryon',z)
densp=results.get_Omega('photon',z)
densn=results.get_Omega('neutrino',z)
densnu=results.get_Omega('nu',z)
densde=results.get_Omega('de',z)
```

```python
Desrad=densnu+densp+densn
Desmat=densb+denscdm
```

```python
plt.plot(a,densK,label='$\Omega_K$')
plt.plot(a,densde,label='$\Omega_{DE}$')
plt.plot(a,Desmat,label='$\Omega_m$')
plt.plot(a,Desrad,label='$\Omega_r$')
plt.xlabel('a(t)')
plt.ylabel('$\Omega$')
plt.legend() 
```

```python
plt.semilogx(a,densK,label='$\Omega_K$')
plt.semilogx(a,densde,label='$\Omega_{DE}$')
plt.semilogx(a,Desmat,label='$\Omega_m$')
plt.semilogx(a,Desrad,label='$\Omega_r$')
plt.xlabel('a(t)')
plt.ylabel('$\Omega$')
plt.legend() 
```
